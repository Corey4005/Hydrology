```python
F_converter = lambda x: (x - 32) * 5 / 9 
```

```python
Julian['2-m Temperature (Celcius)'] = Julian['2-m Temperature (F)'].apply(F_converter)
```

```python
Radiation_converter = lambda x: x * 0.0864
```

```python
Julian['Solar Radiation (W/M^2/D^-1)'] = Julian['Solar Radiation (W/m^2)'].apply(Radiation_converter)
```

```python
Julian
```

```python
import numpy as np
```

```python
wind_speed_corrector = lambda x: x * ((4.87) / np.log((67.8*10)-5.42))
```

```python
Julian['Wind Speed 2m Corrected'] = Julian['Wind Speed at 10m (m/s)'].apply(wind_speed_corrector)
```

```python
Julian
```

```python
import math
```

```python
slope_func = lambda x: math.exp((17.27 * x)/(x + 237.3))*0.6108*4098 / ((x+237.3)**2)
```

```python
Julian['10-m Temp (Celcius)'] = Julian['10-m Temp (F)'].apply(F_converter)
```

```python
Julian
```

```python
Julian['slope'] = Julian['10-m Temp (Celcius)'].apply(slope_func)
```

```python
Julian
```

```python
pressure_converter = lambda x: x*0.1
```

```python
Julian['Pressure - (kPa)'] = Julian['Pressure - (mb)'].apply(pressure_converter)
```

```python
Julian
```

```python
def DT_func(slope, windspeed):
    DT = slope/(slope + (0.00065*(1+(0.34*windspeed))))
    return DT
```

```python
Julian['DT'] = Julian[['slope', 'Wind Speed 2m Corrected']].apply(lambda x: DT_func(*x), axis=1)
```

```python
Julian.set_index('julian time', inplace=True)
```

```python
Julian
```

```python
def PT_func(slope, windspeed): 
    PT = 0.00065/(slope + (0.00065*(1+(0.34*windspeed))))
    return PT
```

```python
Julian['PT'] = Julian[['slope', 'Wind Speed 2m Corrected']].apply(lambda x: PT_func(*x), axis=1)
```

```python
Julian
```

```python
def TT_func(mean_temp, windspeed): 
    TT = (900/(mean_temp + 273))*windspeed
    return TT
```

```python
Julian['TT'] = Julian[['10-m Temp (Celcius)', 'Wind Speed 2m Corrected']].apply(lambda x: TT_func(*x), axis=1)
```

```python
Julian
```

```python
Vapor_pressure = lambda x: 0.6108*(math.exp((17.27*x)/(x+237.3)))
```

```python
Julian['vapor pressure (kPa)'] = Julian['10-m Temp (Celcius)'].apply(Vapor_pressure)
```

```python
Julian
```

```python
max_values = df.groupby(df['Julian Day']).max()
min_values = df.groupby(df['Julian Day']).min()
max_values
min_values
```

```python
Julian['tmax'] = max_values['10-m Temp (F)']
```

```python
Julian.reset_index(inplace=True)
```

```python
max_values.reset_index(inplace=True)
```

```python
min_values.reset_index(inplace=True)
```

```python
Julian['tmax'] = max_values['10-m Temp (F)'].apply(F_converter)
```

```python
Julian['tmin'] = min_values['10-m Temp (F)'].apply(F_converter)
```

```python
E_tmax_func = lambda x: 0.6108*math.exp((17.27*x)/(x+237.3))
E_tmin_func = lambda x: 0.6108*math.exp((17.27*x)/(x+237.3))
```

```python
Julian['e_tmax'] = Julian['tmax'].apply(E_tmax_func)
Julian['e_tmin'] = Julian['tmin'].apply(E_tmin_func)
```

```python
Julian['RH max'] = max_values['RH (%)']
Julian['RH min'] = min_values['RH (%)']
```

```python
Julian['Mean es'] = (Julian['e_tmax'] + Julian['e_tmin']) / 2
```

```python
def ea_func(et_min, RH_max, et_max, RH_min): 
    EA = ((et_min*(RH_max/100))+(et_max*(RH_min/100)))/2
    return EA
```

```python
Julian
```

```python
Julian['Ea'] = Julian[['e_tmin', 'RH max', 'e_tmax', 'RH min']].apply(lambda x: ea_func(*x), axis=1)
```

```python
Julian
```

```python
dr_func = lambda x: 1 + (0.033*math.cos((2*math.pi/365)*x))
```

```python
Julian['Julian Day'] = max_values['Julian Day']
```

```python
Julian['dr'] = Julian['Julian Day'].apply(dr_func)
```

```python
Julian.columns
```

```python
Solar_Declination_func = lambda x: 0.0409*math.sin(((2*math.pi/365)*x)-1.39)
```

```python
Julian['Solar_Declination'] = Julian['Julian Day'].apply(Solar_Declination_func)
```

```python
Julian['latitude'] = 32.723
```

```python
Julian['longitude'] = 86.642
```

```python
Julian['elevation Feet'] = 686
```

```python
def ss_hour_angle(lat, solar_declination):
    angle = np.arccos((-math.tan(lat)*math.tan(solar_declination)))
    return angle
```

```python
Julian['SS hour angle'] = Julian[['latitude', 'Solar_Declination']].apply(lambda x: ss_hour_angle(*x),axis=1)
```

```python
Julian
```

```python
def extraterrestrial_radiation(dr, lat, solar_declination, sunset_hour_angle):
    RA = ((24*60)/math.pi)*(0.0820*dr)*((sunset_hour_angle*math.sin(lat)*math.sin(solar_declination)+math.cos(lat)*math.cos(solar_declination)*math.sin(solar_declination)))
    return RA
```

```python
Julian['RA'] = Julian[['dr', 'latitude', 'Solar_Declination', 'SS hour angle']].apply(lambda x: extraterrestrial_radiation(*x), axis=1)
```

```python
Julian
```

```python
def clear_sky_Rso_func(elevation, Ra):
    RSO = (0.75+((2e10*-5)*elevation))*Ra
    return RSO
```

```python
Julian['RSO'] = Julian[['elevation Feet', 'RA']].apply(lambda x: clear_sky_Rso_func(*x), axis=1)
```

```python
Julian
```

```python
Rns_function = lambda x: (1-0.23)*x
```

```python
Julian['Rns'] = Julian['Solar Radiation (W/M^2/D^-1)'].apply(Rns_function)
```

```python
Julian
```

```python
def Longwave_Radiation_func(Tmax, Tmin, Ea, Rs, Rso):
    RNL = 4.903e-9*((Tmax+273.16)**4 + (Tmin+273.16)**4/2)*(0.34-(0.14*math.sqrt(Ea)))*((1.35*(Rs/Rso))-0.35)
    return RNL
```

```python
Julian.columns
```

```python
Julian['RNL'] = Julian[['tmax', 'tmin', 'Ea', 'Solar Radiation (W/M^2/D^-1)', 'RSO']].apply(lambda x: Longwave_Radiation_func(*x), axis=1)
```

```python
Julian
```

```python
Julian['Net Radiation'] = Julian['Rns'] + Julian['RNL']
```

```python
Julian
```

```python
Julian['Rng'] = 0.408 * Julian['Net Radiation']
```

```python
Julian.set_index(Julian['julian time'], inplace=True)
```

```python
Julian['ETrad'] = Julian['DT']*Julian['Rng']
```

```python
Julian['ETwind'] = Julian['PT']*Julian['TT']*Julian['Ea']
```

```python
Julian['ETo'] = Julian['ETwind'] + Julian['ETrad']
```

```python
import matplotlib.pyplot as plt
Julian.index = pd.to_datetime(Julian.index, utc=True)
```

```python
fig, ax = plt.subplots()
ax.plot(Julian['ETo'], label='P-M method')
ax.set_xticklabels(Julian.index.day)
ax.set_ylabel('Evapotranspiration mm/day')
ax.set_xlabel('Day (June 2012)')
plt.title('Evapotranspiration P-M')
plt.legend()
```

```python
Julian.columns
```

```python
def P_T_function(slope, Net_radiation):
    PETpt = (1.26*slope*(Net_radiation))/((1000*2.45)*(slope+0.066))
    return PETpt
```

```python
Julian['P_T'] = Julian[['slope', 'Net Radiation']].apply(lambda x: P_T_function(*x), axis=1)
```

```python
Julian['P_T'] = Julian['P_T']*1000
```

```python
Julian['P_T']
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(Julian['ETo'], label='P-M method')
ax2.plot(Julian['P_T'], label='P-T method')
ax2.set_xticklabels(Julian.index.day)
ax2.set_ylabel('Evapotranspiration mm/day')
ax2.set_xlabel('Day (June 2012)')
plt.title('Evapotranspiration P-M vs P-T')
plt.legend()
```

The P-M vs P-T methods for June of 2012 shows a phenomenal relationship. As the P-M method moves up and down, the P-T method seems to follow. However, interstingly the P-T method seems to underestimate Evapotranspiration when compared to the P-M method.
